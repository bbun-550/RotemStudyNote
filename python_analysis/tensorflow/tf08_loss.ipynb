```python
# 선형회귀모델 추세식 계산
import numpy as np

class linearRegressionTest:
    def __init__(self, learning_rate, epochs):
        self.w = None
        self.b = None
        self.learning_rate = learning_rate
        self.epochs = epochs
        
    def fit(self, x:np.ndarray, y:np.ndarray):
        # 경사하강법(Gradient Descent)으로 w,b를 학습
        # parameter 초기화
        self.w = np.random.uniform(-2, 2)
        self.b = np.random.uniform(-2, 2)
        
        n = len(x)
        
        for epoch in range(self.epochs):
            y_pred = self.w * x + self.b # 예측값
            loss = np.mean((y - y_pred) ** 2) # 손실
            
            dw = (-2/n) * np.sum(x * (y - y_pred)) # 경사 계산, 편미분
            db = (-2/n) * np.sum(y - y_pred)
            self.w = self.w - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db
            
            # 학습 상태 출력
            if (epoch + 1) % 10 == 0: # 10번에 한번씩만 출력
                print(f'epoch {epoch + 1} / {self.epochs} - loss: {loss:.5f}, w: {self.w:.5f}, b: {self.b:.5f}')
    
    def predict(self, x:np.ndarray):
        if self.w is None or self.b is None:
            raise ValueError("Model parameters w and b are not initialized. Call fit() before predict().")
        return self.w * x + self.b
    
def main():
    np.random.seed(42)
    
    x_height = np.random.normal(175, 10, 30)
    true_w = 0.7
    true_b = -55
    noise = np.random.normal(0,5,30)
    
    # label
    y_weight = true_w * x_height + true_b + noise
    
    print(x_height)
    print(y_weight)
    
    # scaling(표준화)
    x_mean = np.mean(x_height)
    x_std = np.std(x_height)    
    y_mean = np.mean(y_weight)
    y_std = np.std(y_weight)
    
    x_height_scaled = (x_height - x_mean)/x_std
    y_weight_scaled = (y_weight - y_mean)/y_std
    
    # 모델 학습
    model = linearRegressionTest(learning_rate=0.001, epochs=1000)
    model.fit(x_height_scaled, y_weight_scaled)
    
    # 예측
    y_pred_scaled = model.predict(x_height_scaled)
    
    # 예측 결과 역변환 : 표준화했기 때문
    y_pred = (y_pred_scaled * y_std) + y_mean
    print(f'y_pred : {y_pred[:5]}')
    
    # 모델 성능(MSE, R2 → regression) 계산
    mse = np.mean((y_weight - y_pred) ** 2) # R2 = 1 - SSR/SST
    ss_tot = np.sum((y_weight - np.mean(y_weight)) ** 2)
    ss_res = np.sum((y_weight - y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    
    print('학습결과 ------------')
    print(f'추정된 w 기울기 : {model.w:.4f}')
    print(f'추정된 b 절편 : {model.b:.4f}')
    
    for i in range(len(x_height)):
        print(f'키 : {x_height[i]:.2f}cm, 몸무게 실제: {y_weight[i]:.2f}kg, 예측 : {y_pred[i]:.2f}kg')
    
    print(f'mse : {mse:.5f}')
    print(f'r2 : {r2:.5f}')

if __name__ == '__main__':
    main()
```
